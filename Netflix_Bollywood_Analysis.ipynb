```python
import numpy as np
import pandas as pd
import plotly.express as px
```

### Conversion of duration columns dtypes to datetime dtypes and spliting the year, month and date columns

```python
netflix_data['new_date'] = pd.to_datetime(netflix_data['date_added'], errors= 'coerce')
netflix_data['format_date'] = netflix_data['new_date'].dt.strftime('%Y/%B/%d')
netflix_data[['year', 'month', 'date']] = netflix_data['format_date'].str.split('/', expand= True)
netflix_data = netflix_data.drop(['new_date', 'date_added'], axis= 1)
netflix_data
```

```python
netflix_data.head()
```

### Filling the int, object dtypes values on places of NULL values

```python
int_columns = netflix_data.select_dtypes('int').columns
object_columns = netflix_data.select_dtypes('object').columns

for i in netflix_data.columns:
    if i in int_columns:
        netflix_data[i] = netflix_data[i].fillna(0)
    if i in object_columns:
        netflix_data[i] = netflix_data[i].fillna('Not Available')
```

### Find out which types of genre Movies and Tv Shows are with most number

```python
value = ['Movie', 'TV Show']
results = {}

for i in value:
    filter_listed_data = netflix_data[~netflix_data['listed_in'].str.contains('Not Available') & (netflix_data['type'] == i)]
    listed_in_movie = filter_listed_data['listed_in'].str.split(',').explode().str.strip().value_counts().reset_index().head(10)
    results[i] = listed_in_movie

data = pd.concat(results, axis=1)

data.columns = ['Movie_Listed_In', 'Movie_Count', 'TV_Show_Listed_In', 'TV_Show_Count']
for col in data.columns:
    if 'Movie_Count' in col:
        fig = px.bar(data, x='Movie_Listed_In', y='Movie_Count', title='Top 10 Genres in Movies')
        fig.update_traces(marker_color= 'pink', textposition= 'outside', texttemplate= '%{y}', textfont= dict(color= 'pink'))
        fig.update_layout(xaxis= dict(title= 'Movies Genrs', showgrid= False),
                         yaxis= dict(title= 'Count of Genres', showgrid= False),
                         plot_bgcolor='Turquoise',
                         title_font= dict(size= 20))
        fig.show()

    if 'TV_Show_Count' in col:
        fig = px.bar(data, x='TV_Show_Listed_In', y='TV_Show_Count', title='Top 10 Genres in TV Shows')
        fig.update_traces(marker_color= 'red', textposition= 'outside', texttemplate= '%{y}', textfont= dict(color= 'red'))
        fig.update_layout(xaxis= dict(title= 'TV Shows Genres', showgrid= False),
                         yaxis= dict(title= 'Count of Genres', showgrid=False),
                         plot_bgcolor= 'green',
                         title_font= dict(size= 20))
        fig.show()
```
